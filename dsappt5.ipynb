```python
1. 
 def convert_to_2d_array(original, m, n):
    if len(original) != m * n:
        return []
    
    result = [[0] * n for _ in range(m)]
    
    for i in range(len(original)):
        row = i // n
        col = i % n
        result[row][col] = original[i]
    
    return result

##using the given example

original = [1, 2, 3, 4]
m = 2
n = 2

print(convert_to_2d_array(original, m, n))

#output

[[1, 2], [3, 4]]



2.
def count_complete_rows(n):
    complete_rows = 0
    total_coins = 0

    for i in range(1, n + 1):
        total_coins += i
        if total_coins <= n:
            complete_rows += 1

    return complete_rows


n = 5
print(count_complete_rows(n))

#output
2



3.
def sorted_squares(nums):
    n = len(nums)
    result = [0] * n
    left = 0
    right = n - 1

    for i in range(n - 1, -1, -1):
        if abs(nums[left]) >= abs(nums[right]):
            result[i] = nums[left] ** 2
            left += 1
        else:
            result[i] = nums[right] ** 2
            right -= 1

    return result

nums = [-4, -1, 0, 3, 10]
print(sorted_squares(nums))

#output

[0, 1, 9, 16, 100]



4.
def find_disjoint(nums1, nums2):
    set_nums1 = set(nums1)
    set_nums2 = set(nums2)

    nums1_not_in_nums2 = list(set_nums1 - set_nums2)
    nums2_not_in_nums1 = list(set_nums2 - set_nums1)

    return [nums1_not_in_nums2, nums2_not_in_nums1]


nums1 = [1, 2, 3]
nums2 = [2, 4, 6]
print(find_disjoint(nums1, nums2))

#output

[[1, 3], [4, 6]]



5.
def find_distance_value(arr1, arr2, d):
    distance = 0

    for num1 in arr1:
        found = True

        for num2 in arr2:
            if abs(num1 - num2) <= d:
                found = False
                break

        if found:
            distance += 1

    return distance


arr1 = [4, 5, 8]
arr2 = [10, 9, 1, 8]
d = 2
print(find_distance_value(arr1, arr2, d))

#output

2


6.
def find_duplicates(nums):
    n = len(nums)
    result = []

    for num in nums:
        index = abs(num) - 1
        if nums[index] > 0:
            nums[index] = -nums[index]
        else:
            result.append(abs(num))

    return result

nums = [4, 3, 2, 7, 8, 2, 3, 1]
print(find_duplicates(nums))

#output

[2, 3]



7.
def find_minimum(nums):
    n = len(nums)
    left = 0
    right = n - 1

    while left < right:
        mid = (left + right) // 2

        if nums[mid] > nums[right]:
            left = mid + 1
        else:
            right = mid

    return nums[left]

nums = [3, 4, 5, 1, 2]
print(find_minimum(nums))


#output
1



8.
from collections import defaultdict

def findOriginalArray(changed):
    if len(changed) % 2 != 0:
        return []  # If the changed array has an odd length, it can't be a doubled array
    
    counts = defaultdict(int)
    for num in changed:
        counts[num] += 1
    
    original = []
    for num in changed:
        if counts[num] > 0:
            if counts[num * 2] > 0:
                counts[num * 2] -= 1
                original.append(num // 2)
            else:
                return []  # If the doubled value doesn't exist in the HashMap or its count is 0, it's not a doubled array.
    
    return original



changed = [1, 3, 4, 2, 6, 8]
print(findOriginalArray(changed))


#output
[1, 3, 4]


```
